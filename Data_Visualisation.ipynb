This notebook is provide some insights about a company by looking at its former employees' data. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_excel("/content/data.xlsx")
```

```python
mba = {'E1': 0, 'M1': 0, 'E2': 0, 'M3': 0, 'M4': 0, 'M2': 0, 'CXO': 0}
bachelors = {'E1': 0, 'M1': 0, 'E2': 0, 'M3': 0, 'M4': 0, 'M2': 0, 'CXO': 0}
grades = ['E1','E2','M1','M2','M3','M4','CX0']
bachnum = 0
mbanum = 0
bar = np.arange(len(grades))
bar2 = [i+0.2 for i in bar]
bar1 = [i-0.2 for i in bar]

for index in df.index:
  if df['Education'][index] == 'MBA':
    mbanum += 1
    mba[df['Grade'][index]] += 1
  else:
    bachnum += 1
    bachelors[df['Grade'][index]] += 1
for i in bachelors:
  bachelors[i] *= 100/bachnum
for i in mba:
  mba[i] *= 100/mbanum
plt.bar(bar1, list(mba.values()),0.4)
plt.bar(bar2, list(bachelors.values()),0.4)
plt.xticks(bar1, list(bachelors.keys()))
plt.legend(['MBA','Bachelors'], fontsize="x-large")
plt.xlabel('Grades')
plt.ylabel('Percentage')


    
```

It can be seen that MBA graduates  have a better grade, with the majority at M2 grade than the Undergraduates, which majorly have an E1 grade, hence it can be predicted that this company favours managerial expertise strongly.

```python
Male = []
Female = []
for index in df.index:
  if df['Gender'][index] == 'Male':
    Male.append(int(df['Monthly Income'][index]))
  else:
    Female.append(int(df['Monthly Income'][index]))
    
```

```python
plt.hist(Male,10,(25000,100000), alpha = 0.5)
plt.xlabel('Monthly income')
plt.ylabel('Number of Male employees')
```

```python
plt.hist(Female,10,(25000,100000), alpha = 0.4)
plt.xlabel('Monthly Income')
plt.ylabel('Number of female employees')
```

It can be observed that the general trends show that there is not much distinction between the salarieds offered to male and female employees. We can predict that the head of the company or the H is a woman and is feminist. Though it is still not able to encourage many woman to join the workforce as can be seen by the difference in the number of male and female employees

```python
remarks = {'Issues with the Manager': 0,'More Challenging Job Roles/ Higher designation': 0,'Lack of Growth' : 0,'Better Salary' : 0,'Absconding' : 0, 'Others' : 0}
for index in df.index:
  t = df['Remarks'][index]
  if t in remarks:
    remarks[t] += 1
  else:
    remarks['Others'] += 1
plt.pie(remarks.values(),labels = remarks.keys())  
plt.title('Reasons for leaving the company')
```

From the graph it is evident that the main reason for people leaving the comapny is issues with the manager , followed by lack of growth.
Thus, the manager needs to be replaced and the work culture shouls be employee friendly like Analytics Club, IIT Madras, Chennai. :)



```python
Tenures = []
for index in df.index:
  Tenures.append(float(df['Tenure'][index].replace(' ','')))

```

```python
plt.hist(Tenures, 50)
plt.xlabel('Years')
plt.ylabel('Number of employees')
```

It can be observed that there is a steady decrease in the number of employees takes place as the tenure increases. This shows that the employees are not very happy with staying out into the company, and join it possibly only to earn their bread while searching for some other place to work at.

```python
#Gender - Tenure
M,F = [] , []
T = df["Tenure"]
c = 0
def norm(s):
  snew = ''
  for i in s:
    if i != ' ':
      snew += i
  return float(snew)
for i in df["Gender"]:
  if (i == "Male"):
    M.append(norm(T[c]))
  else:
    F.append(norm(T[c]))
  c += 1

fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(M, bins = [1,2,3,4,5,6,7])
plt.title("Tenure duration of men ")
plt.xlabel('Years')
plt.ylabel('Number of employees')
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(F, bins = [1,2,3,4,5,6,7])
plt.title("Tenure duration of women ")
plt.xlabel('Years')
plt.ylabel('Number of employees')
# Show plot
plt.show()
```

On an average scale , women tend to stay in the company for a longer duration , while many men tend to leave early. but the people who have stayed the longest in the comapny are men.

```python

M,F = [] , []
T = df["Monthly Income"]
A = [0]*7
B = [0]*7
c = 0
for i in df["Age"]:
  A[(i-20)//5] += T[c]
  B[(i-20)//5] += 1
  c += 1
for i in range(len(A)):
  A[i] = A[i]/B[i]
  
fig = plt.figure(figsize = (10, 5))
C = []
for i in range(len(A)):
  C.append(i+1)
plt.bar(C, A, color ='blue',
        width = 0.4)
plt.xlabel("Age of employees")
plt.ylabel("Average income")
plt.title("Average salary across different age groups")
plt.show()
for i in range(len(A)):
  print(i+1 , "Age group" , i*5 + 20 , "to" , i*5 + 25)
```

It can be seen that as age increases, the average salary increases, thus we can conclude that the company is seeking for more experienced people for the job roles.
